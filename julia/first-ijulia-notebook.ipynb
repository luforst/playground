## Broadcasting

```julia
function f(x)
    x^2
end
A = [i + 3*j for j in 0:2, i in 1:3]
```

```julia
f(A)
```

```julia
f.(A) # this is broadcasting: f(x) acts upon the singular elements of the matrix instead of on the matrix itself
```

## Structs
structs allow us to create custom types, similar to classes in OOP languages

```julia
struct MyObj
    field1
    field2
end
```

```julia
myobj1 = MyObj("hello", "world")
display(myobj1)
myobj1.field1

myobj1.field1 = "test" # impossible because it is an immutable struct
```

```julia
mutable struct Person # now we have a mutable struct
    name::String
    age::Float64
end
```

```julia
irrwitzer = Person("Beelzebub Irrwitzer", 187)
```

```julia
irrwitzer.age += 1
```

```julia
mutable struct Person
    name::String
    age::Float64
    isActive

    function Person(name, age) #constructor
        new(name, age, true)
    end
end
```

```julia
irrwitzer = Person("Beelzebub Irrwitzer", 187)
```

```julia
function birthday(person::Person)
    person.age += 1
end
```

```julia
birthday(irrwitzer)
```
